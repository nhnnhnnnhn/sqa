```python
import csv
import re

input_file = "src_data/all_labeled_questions_1.csv"
output_file = "src_data/all_labeled_questions_2.csv"

pattern = r'([A-D]\.\s.*?)(?=\s[A-D]\.|$)'

with open(input_file, newline='', encoding='utf-8') as f_in, \
     open(output_file, 'w', newline='', encoding='utf-8') as f_out:

    reader = csv.reader(f_in)
    writer = csv.writer(f_out)

    for row in reader:
        if len(row) < 2:
            continue

        label, content = row[0], row[1]

        if label == "answer":
            # Tách các đáp án A. B. C. D.
            matches = re.findall(pattern, content)
            if matches:
                for ans in matches:
                    ans = ans.strip()
                    if ans:
                        writer.writerow(["answer", ans])
            else:
                # answer đã chuẩn → giữ nguyên
                writer.writerow([label, content])
        else:
            # question → giữ nguyên
            writer.writerow([label, content])

print("✅ Đã tách đáp án xong!")
```
